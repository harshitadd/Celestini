```
import numpy as np 
import matplotlib.pyplot as plt 
import csv
```

```
airveda = []
k=0
with open ('/content/airveda_data__181920.csv') as file:
  reader = csv.reader(file, delimiter  = ',')
  for row in reader:
    k+=1
    if(k>1):
      try:
        l=[]
        l.append(row[0])
        temp = row[0]
        temp=temp.split('-')
        newtemp=temp[2].split(' ')
        newesttemp=newtemp[1].split(':')
        j=int(newesttemp[1])
        if(j<=7.5):
          final = newtemp[0] + '-' + temp[1] + '-' + temp[0] + ' ' + newesttemp[0] + ':' + '00'
        elif(j<=15 or j<=22.5):
          final = newtemp[0] + '-' + temp[1] + '-' + temp[0] + ' ' + newesttemp[0] + ':' + '15'
        elif(j<=30 or j<=37.5):
          final = newtemp[0] + '-' + temp[1] + '-' + temp[0] + ' ' + newesttemp[0] + ':' + '30'
        elif(j<=45 or j<=52.5):
          final = newtemp[0] + '-' + temp[1] + '-' + temp[0] + ' ' + newesttemp[0] + ':' + '45'
        elif(j>52.5):
          final = newtemp[0] + '-' + temp[1] + '-' + temp[0] + ' ' + newesttemp[0] + ':' + '00'
        print(final)
        l.append(final)
        ##l.append(row[2])
        l.append(row[3])
        airveda.append(l)
      except: 
        print('error')
        print(temp)
```
