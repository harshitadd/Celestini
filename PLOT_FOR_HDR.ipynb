```
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```
img_list = ['FJIMG_20190613_125630.jpg','FJIMG_20190613_125638.jpg','FJIMG_20190613_125645.jpg','FJIMG_20190613_125652.jpg']
images=[]
for im in img_list:
    images.append(cv.imread(im))

exposure_times = np.array([1/1250.0,1/320.0,1/5140.0,1/8411.0], dtype=np.float32) 
#print(np.shape(images))


cal_debevec = cv.createCalibrateDebevec()
crf_debevec = cal_debevec.process(images,times=exposure_times)

# print(cal_debevec)
# print(crf_debevec)

#or x in np.nditer(crf_debevec):
#   print(x)
#print(np.shape(crf_debevec))

# cv.normalize(crf_debevec,crf_debevec,0,255)
# #print(crf_debevec)


p = np.arange(0,256)
b=[]
g=[]
r=[]
for i in range(0,256,1):
    b.append(crf_debevec[i][0][0])
    g.append(crf_debevec[i][0][1])
    r.append(crf_debevec[i][0][2])

    
plt.plot(p,b)
plt.plot(p,g)
plt.plot(p,r)

f , ax = plt.subplots(1,4,figsize = (30,10))
ax[0].plot(p,b)
ax[1].plot(p,g)
ax[2].plot(p,g)

channel = [b,g,r]

for i in channel:
    ax[3].plot(p,i)

# print(max(b),min(b))
# print(max(g),min(g))
# print(max(r),min(r))
```
