```
import csv 
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.preprocessing import MinMaxScaler 
```

```
cpcb_time={}
k=0
with open('/content/cpcb_yearly.csv') as cpcb: 
  reader=csv.reader(cpcb, delimiter = ',')
  for row in reader:
    k+=1
    if(k>=18):
      dummy=row[0]
      dummy=dummy.split('-')
      temp = dummy[2].split(' ')
      n = temp[1].split(':')
      dummy = str(dummy[0])+str(dummy[1])+str(temp[0]) + str(n[0]) + str(n[1])
      cpcb_time[dummy]=row[2]   

      
month_31=np.arange(1,32,1)
month_30=np.arange(1,31,1)
feb_time=np.arange(1,29,1)


january=[]
feb=[]
march=[]
april=[]
may=[]
june=[]
july=[]
aug=[]
sept=[]
october=[]
nov=[]
dec=[]

ctr=0
for key in cpcb_time:
  ctr+=1
  if(ctr<=31):
    if(cpcb_time[key]!='None'):
      january.append(cpcb_time[key])
    else:
      january.append('NAN')
  elif(ctr<=59):
      if(cpcb_time[key]!='None'):
        feb.append(cpcb_time[key])
      else:
        feb.append('NAN')
  elif(ctr<=90):
      if(cpcb_time[key]!='None'):
        march.append(cpcb_time[key])
      else:
        march.append('NAN')
  elif(ctr<=120):
      if(cpcb_time[key]!='None'):
        april.append(cpcb_time[key])
      else:
        april.append('NAN')
  elif(ctr<=151):
      if(cpcb_time[key]!='None'):
        may.append(cpcb_time[key])
      else:
        may.append('NAN')
  elif(ctr<=181):
      if(cpcb_time[key]!='None'):
        june.append(cpcb_time[key])
      else:
        june.append('NAN')
  elif(ctr<=212):
      if(cpcb_time[key]!='None'):
        july.append(cpcb_time[key])
      else:
        july.append('NAN')
  elif(ctr<=243):
      if(cpcb_time[key]!='None'):
        aug.append(cpcb_time[key])
      else:
        aug.append('NAN')
  elif(ctr<=273):
      if(cpcb_time[key]!='None'):
        sept.append(cpcb_time[key])
      else:
        sept.append('NAN')
  elif(ctr<=304):
      if(cpcb_time[key]!='None'):
        october.append(cpcb_time[key])
      else:
        october.append('NAN')
  elif(ctr<=334):
      if(cpcb_time[key]!='None'):
        nov.append(cpcb_time[key])
      else:
        nov.append('NAN')
  elif(ctr<=365):
      if(cpcb_time[key]!='None'):
        dec.append(cpcb_time[key])
      else:
        dec.append('NAN')


```

```
nov = list(map(float,nov))
sept = list(map(float,sept))
october = list(map(float,october))
aug = list(map(float,aug))
july = list(map(float,july))
dec = list(map(float,dec))
january = list(map(float,january))
feb = list(map(float,feb))
march = list(map(float,march))
april= list(map(float,april))
may = list(map(float,may))
june = list(map(float,june))


f, ax = plt.subplots(2, figsize=(20,20))

ax[0].plot(month_31,january)                               #MAX - 100 
ax[0].plot(feb_time,feb)                                   #MAX - 120 
ax[0].plot(month_31,march)                                 #MAX - 300
ax[0].plot(month_30,april)                                 #MAX - 600 
ax[0].plot(month_31,may)
ax[0].plot(month_30,june)
ax[0].plot(month_31,july)                                  #MAX - 100 
ax[0].plot(month_31,aug)                                   #MAX - 120
ax[0].plot(month_30,sept)                                  #MAX - 120
ax[0].plot(month_31,october)                               #MAX - 300
ax[0].plot(month_30,nov)                                   #MAX - 600 
ax[0].plot(month_31,dec)                                   #MAX - 400 
```
