```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img_list = ['/home/harshita/Celestini/CRF/image1.JPG','/home/harshita/Celestini/CRF/image2.JPG','/home/harshita/Celestini/CRF/image3.JPG','/home/harshita/Celestini/CRF/image4.JPG']
images=[]
for im in img_list:
    images.append(cv.imread(im))

exposure_times = np.array([0.0333, 0.25, 2.5,15.0], dtype=np.float32) 
#print(np.shape(images))


cal_debevec = cv.createCalibrateDebevec()
crf_debevec = cal_debevec.process(images,times=exposure_times)

# print(cal_debevec)
# print(crf_debevec)

#or x in np.nditer(crf_debevec):
#   print(x)
#print(np.shape(crf_debevec))

p = np.arange(0,256)
b=[]
g=[]
r=[]
for i in range(0,256,1):
    b.append(crf_debevec[i][0][0])
    g.append(crf_debevec[i][0][1])
    r.append(crf_debevec[i][0][2])

f , ax = plt.subplots(1,4,figsize = (30,10))
ax[0].plot(p,b)
ax[1].plot(p,g)
ax[2].plot(p,g)

channel = [b,g,r]

for i in channel:
    ax[3].plot(p,i)
#ax[3].plot(crf_debevec)


#is the pixel values (255) vs the exposure curve 
```

```python

merge = cv.createMergeDebevec()
hdr = merge.process(images, exposure_times, crf_debevec)

# print(hdr)
cv.imwrite('/home/harshita/Celestini/CRF/hdr_image.hdr', hdr)
# plt.imshow(hdr)
# plt.show()
```
