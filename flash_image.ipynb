```
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import minmax_scale 
```

```
user_img=['IMG_20190614_111546.jpg','IMG_20190614_111550.jpg']
images=[]
for im in user_img: 
  images.append(cv.imread(im))
  
exposure_times=np.array([1/50.0,1/33.0],dtype=np.float32)

cdev=cv.createCalibrateDebevec()
response=cdev.process(images, times= exposure_times)

p = np.arange(0,256)
b=[]
g=[]
r=[]
for i in range(0,256,1):
    b.append(response[i][0][0])
    g.append(response[i][0][1])
    r.append(response[i][0][2])
    


b=minmax_scale(b)
g=minmax_scale(g)
r=minmax_scale(r)
p=minmax_scale(p)

channel = [b,g,r]
for i in channel:
    plt.plot(p,i)


cmerge=cv.createMergeDebevec()
hdr= cmerge.process(images,exposure_times, response)

#print(hdr)

# print(max(hdr.flatten()))
# print(min(hdr.flatten()))
# plt.plot(hdr)
# plt.show()
```
